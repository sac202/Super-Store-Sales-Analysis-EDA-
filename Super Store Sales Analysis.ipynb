```python
sns.barplot(x='region', y='profit', data=df, estimator=sum, errorbar=None)
plt.title("Total Profit by Region")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
sns.boxplot(x='category', y='discount', data=df)
plt.title("Discount Range by Category")
plt.show()
```

```python
df['month'] = df['order_date'].dt.to_period('M')
monthly_sales = df.groupby('month')['sales'].sum()

monthly_sales.plot(kind='line', figsize=(10,5), title="Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.show()
```


* Correlation Analysis:

---> Sales and Shipping Cost show a strong positive correlation (0.77), indicating higher sales generally involve higher shipping expenses.

---> Discount and Profit are negatively correlated (-0.32) — suggesting that higher discounts tend to reduce profit margins.

---> Sales and Profit are moderately correlated (0.49), showing that increased sales usually lead to higher profits but not always proportionally.

* Distribution of Sales:

---> The sales distribution is right-skewed, meaning a large number of transactions have low sales values, while few transactions contribute very high sales.

* Profit by Category:

---> All three categories (Office Supplies, Furniture, Technology) show the presence of outliers in profit values, especially in Technology.

* Discount vs Profit Relationship:

---> The scatter plot shows a downward trend, confirming that as discount increases, profit tends to decrease significantly.

* Profit by Region:

---> The Central region recorded the highest total profit, while regions like Canada and Southeast Asia showed relatively lower profits.

*Discount by Category:

---> Furniture receives the highest range of discounts, followed by Office Supplies and Technology.

* Monthly Sales Trend:

---> There’s a clear upward trend in sales from 2011 to 2014, indicating overall business growth over time with seasonal fluctuations.
